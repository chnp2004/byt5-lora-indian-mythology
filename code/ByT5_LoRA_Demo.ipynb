```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_path = "/content/drive/MyDrive/byt5_final_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path).to("cuda")

question = "How did Rama initially react to Supanakha's advances or overtures?"
context = "The prince responded to the demoness's crude advances with humor. Rama, with a sense of humor and mischief, calmly and jokingly rejected her advances, stating that he was already married to Sita. He playfully suggested she approach his unmarried brother, Lakshman."
input_text = f"question: {question} context: {context}"

inputs = tokenizer(input_text, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=512)
print("ðŸ§  Answer:", tokenizer.decode(outputs[0], skip_special_tokens=True))
```
